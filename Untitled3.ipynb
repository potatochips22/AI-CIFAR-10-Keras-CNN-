## Integrated Transfer Learning Model

This section consolidates all the steps involved in setting up, training, and fine-tuning the MobileNetV2-based transfer learning model.

```python
import numpy
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.layers import Activation, BatchNormalization, Conv2D, Dense, Dropout, Flatten, MaxPooling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.applications import MobileNetV2
from tensorflow.python.keras.utils import np_utils

# Fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)

# Load data
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize inputs from 0-255 to 0.0-1.0
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
X_train = X_train / 255.0
X_test = X_test / 255.0

# One hot encode outputs
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]

print("Data loaded and preprocessed.")
```

### 1. Resize Images and Load Pre-trained MobileNetV2 Base

Images are resized to (96, 96, 3) to be compatible with MobileNetV2 and to manage memory usage. The MobileNetV2 convolutional base is loaded with ImageNet weights and its layers are frozen to use it as a feature extractor.

```python
# 1. 定義預處理函數
def preprocess_image(image, label):
    # 在這裡進行 resize，這會在模型訓練時逐批進行
    image = tf.image.resize(image, (96, 96))
    # 如果 MobileNetV2 需要，也可以在這裡做歸一化 (Normalization)
    # image = tf.keras.applications.mobilenet_v2.preprocess_input(image)
    return image, label

# 2. 建立訓練集流水線
train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train_ds = train_ds.shuffle(buffer_size=1000) # 打亂數據
train_ds = train_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE) # 即時處理
train_ds = train_ds.batch(32).prefetch(tf.data.AUTOTUNE) # 緩衝處理，讓 GPU 不用等 CPU

# 3. 建立測試集流水線
test_ds = tf.data.Dataset.from_tensor_slices((X_test, y_test))
test_ds = test_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
test_ds = test_ds.batch(32).prefetch(tf.data.AUTOTUNE)
```

```python
# Resize the input images to 96x96
X_train_resized = tf.image.resize(X_train, (96, 96))
X_test_resized = tf.image.resize(X_test, (96, 96))

# Load the MobileNetV2 convolutional base, excluding the top classification layer
conv_base = MobileNetV2(weights='imagenet', include_top=False, input_shape=(96, 96, 3))

# Freeze the weights of the convolutional base
conv_base.trainable = False

print("MobileNetV2 convolutional base loaded and weights frozen with 96x96 input.")
conv_base.summary()
```

### 2. Build New Classification Head

A custom classification head is added on top of the frozen convolutional base. This head is designed to classify images into 10 CIFAR-10 classes.

```python
# Create a new classification head
classification_head = Sequential([
    Flatten(),
    Dense(512),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

# Combine the frozen convolutional base with the new classification head
model = Sequential([
    conv_base,
    classification_head
])

model.summary()
```

### 3. Compile and Train Model (Feature Extraction)

The model is compiled with the Adam optimizer and trained. Only the newly added classification head's weights are updated in this phase, as the `conv_base` is frozen. A learning rate scheduler is also applied.

```python
epochs = 25 # Number of epochs for initial feature extraction training

# Compile the model with Adam optimizer
model.compile(optimizer=Adam(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Define learning rate scheduler
lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=0.00001,
    verbose=1
)

print(f"Training classification head for {epochs} epochs (feature extraction)...")

# Train the model (feature extraction)
train_history_feature_extraction = model.fit(X_train_resized, y_train,
                                           validation_data=(X_test_resized, y_test),
                                           epochs=epochs,
                                           batch_size=32,
                                           verbose=2,
                                           callbacks=[lr_scheduler])
```

### 4. Unfreeze and Fine-tune Model

Some layers of the convolutional base are unfrozen, and the entire model is then fine-tuned with a very low learning rate to allow for subtle adjustments to the pre-trained weights, potentially improving accuracy further.

```python
conv_base.trainable = True

# Freeze the first 100 layers (or adjust based on experimentation) to fine-tune only deeper layers
for layer in conv_base.layers[:100]:
    layer.trainable = False

print("Convolutional base partially unfrozen for fine-tuning.")
conv_base.summary()

# Recompile the model with a very low learning rate for fine-tuning
fine_tune_epochs = 10 # Number of epochs for fine-tuning

model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(f"Fine-tuning the model for {fine_tune_epochs} epochs...")

# Train the model for fine-tuning
train_history_fine_tune = model.fit(X_train_resized, y_train,
                                           validation_data=(X_test_resized, y_test),
                                           epochs=fine_tune_epochs,
                                           batch_size=32,
                                           verbose=2,
                                           callbacks=[lr_scheduler])
```

### 5. Evaluate Final Model Performance

Evaluation of the fine-tuned model on the test dataset.

```python
scores_fine_tune = model.evaluate(X_test_resized, y_test, verbose=0)
print("Fine-tuned Model Accuracy: %.2f%%" % (scores_fine_tune[1] * 100))
```

### 提升模型準確率的建議

以下是一些提升 CNN 模型在影像分類任務（例如 CIFAR-10 資料集）中準確率的常用技巧：

1. **資料增強**：CIFAR-10 資料集相對較小。透過對影像應用隨機變換（例如水平翻轉、隨機裁剪、旋轉、平移、亮度調整）來人為地增加訓練資料集的大小，可以顯著防止過度擬合並提高模型的泛化能力。 Keras 提供了 `ImageDataGenerator` 來實現這一目的。

2. **批量歸一化**：在卷積層或全連接層之後添加批量歸一化層（`tf.keras.layers.BatchNormalization`）可以幫助穩定並加速訓練過程。它對前一層的活化值進行歸一化，從而減少內部協變量偏移。

3. **學習率調度**：使用學習率調度而非固定學習率，可以幫助優化器更有效地收斂。例如，可以使用 `ReduceLROnPlateau`、`CosineDecay` 或自訂調度。這有助於在損失函數中找到更優的最小值。

4. **更高級的優化器**：雖然帶動量的隨機梯度下降 (SGD) 效果不錯，但嘗試其他自適應優化器，例如 Adam、RMSprop 或 Adagrad，可能會帶來更快的收斂速度和更高的最終精度。

5. **更深層更複雜的網路架構**：目前模型相對較淺。探索具有更多層數（例如，更多卷積塊、更大濾波器尺寸）的架構，或使用 ResNet、VGG 或 Inception（即使是較小的版本）等先進架構，可以捕捉更複雜的特徵。但是，請注意計算成本的增加以及過度擬合的風險。

6. **遷移學習**：對於影像分類，使用在 ImageNet 等大型資料集上預先訓練的模型（例如 VGG16、ResNet50、MobileNetV2），然後針對 CIFAR-10 進行微調，通常是最有效的方法。您可以將預訓練模型用作特徵提取器，也可以微調其頂層。

7. **正則化（超越 Dropout）**：雖然您已經使用了 Dropout，但還可以向卷積層和全連接層添加其他形式的正則化，例如 L1/L2 權重正則化，以懲罰過大的權重並進一步防止過擬合。

8. **超參數調優**：系統地探索不同的超參數組合（例如，epoch 數、批次大小、學習率、dropout 率、濾波器數量、內核大小），可以使用網格搜尋或隨機搜尋等技術，或者使用貝葉斯優化等更高級的方法，可以找到性能更好的配置。
